## Install Required Libraries

```python

# Import Libraries
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
from census import Census
from us import states
import censusdata
```

## Set the Census API Key

```python

# Replace 'YOUR_API_KEY' with your actual API key obtained from http://api.census.gov/data/key_signup.html
c = Census("YOUR_API_KEY")

# Exploring available ACS Variables
pd.DataFrame(c.acs5.tables())

```

## Next, retrieve Data. 

Let's start with an example of Median Household Income for New York City

```python

# Get median household income for NYC Census tracts
nyc_income = c.acs5.state_county_tract(
    fields=["B19013_001E"],  # Median household income variable
    state_fips=states.NY.fips,
    county_fips="061",       # New York County (Manhattan)
    tract="*",
    year=2022
)

# Convert to DataFrame
nyc_income_df = pd.DataFrame(nyc_income)
```

Download geometries

```python
# Download the geometries for Census tracts in the state of New York (state code: 36) - note the 36 in the url
nyc_tracts = gpd.read_file(
    "https://www2.census.gov/geo/tiger/TIGER2022/TRACT/tl_2022_36_tract.zip"
)

# # Filter to include only New York County (county code "061")
nyc_tracts = nyc_tracts[nyc_tracts["COUNTYFP"] == "061"]
```

Create a new column in income dataset to match the 'GEOID' keys in geometries to then merge the data

```python
nyc_income_df['GEOID'] = [nyc_income_df.loc[i, 'state'] + nyc_income_df.loc[i, 'county'] + 
                         nyc_income_df.loc[i, 'tract'] for i in range(len(nyc_income_df))]
```

```python
nyc_income_df = pd.merge(nyc_income_df, nyc_tracts[['GEOID', 'geometry']], how='left', on='GEOID')
```

Convert merged data to geodataframe

```python
nyc_income_gdf = gpd.GeoDataFrame(nyc_income_df, geometry='geometry')
```

Keep only valid values of income variable

```python
nyc_income_gdf = nyc_income_gdf[nyc_income_gdf['B19013_001E'] >= 0]
```

Make a plot

```python
fig, ax = plt.subplots(figsize=(10, 10))
nyc_income_gdf.plot(column="B19013_001E", cmap="magma", legend=True, ax=ax)
ax.set_title("Median Household Income in New York City (2022)")
plt.show()
```

Example - Race

```python
# Step 7: Retrieve Data - Median Household Income for Los Angeles County

# Get black race for LA Census tracts
la_race = c.acs5.state_county_tract(
    fields=[
        "B02001_003E",  # Black or African American Alone
    ],
    state_fips=states.CA.fips,
    county_fips="037",  # Los Angeles County
    tract="*",
    year=2022
)

# Convert to DataFrame
la_race_df = pd.DataFrame(la_race)
```

```python
# Download the geometries for Census tracts in the state of California (state code: 06) - note the 06 in the url
la_tracts = gpd.read_file(
    "https://www2.census.gov/geo/tiger/TIGER2022/TRACT/tl_2022_06_tract.zip"
)

# # Filter to include only New York County (county code "037")
la_tracts = nyc_tracts[nyc_tracts["COUNTYFP"] == "037"]
```

```python
# Create a new column in income dataset to match the 'GEOID' keys in geometries to then merge the data
la_race_df['GEOID'] = [la_race_df.loc[i, 'state'] + la_race_df.loc[i, 'county'] + 
                         la_race_df.loc[i, 'tract'] for i in range(len(la_race_df))]
```

```python
la_race_df = pd.merge(la_race_df, la_tracts[['GEOID', 'geometry']], how='left', on='GEOID')

la_race_gdf = gpd.GeoDataFrame(la_race_df, geometry='geometry')
```

```python
# Keep only valid values

la_race_gdf = la_race_gdf[la_race_gdf['B02001_003E'] >= 0]
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
la_race_gdf.plot(column="B02001_003E", cmap="magma", legend=True, ax=ax)
ax.set_title("Total population count (race: black)")
plt.show()
```
